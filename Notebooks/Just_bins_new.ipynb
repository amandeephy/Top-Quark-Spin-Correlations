```python
# Function to calculate the exponential with constants a and b
def exponential(x, start, hl):
    return scipy.stats.expon.pdf(x, loc = start , scale = hl)

def gaussian(x, mean, std):
    return scipy.stats.norm.pdf(x, loc = mean , scale = std)

def total(x, a, b, mean, amp, std_dev):
    return (exponential(x, a, b) + gaussian(x, mean, amp, std_dev))
```

```python
def scp(x, *args) :
    mean, std, gamp, start, hl, eamp = args # mean, std dev, gaussian and expo amplitudes and half life
    ret  = 0
    ret += gamp * scipy.stats.norm.pdf(x , loc = mean , scale = std)
    ret += (1 - gamp) * scipy.stats.expon.pdf(x, loc = start , scale = hl)
    return ret
```

```python
def landau(x, *args) :
    start, hl = args
    return scipy.stats.moyal.pdf(x, loc = start, scale = hl)
```

```python
def custom(x, *args) :
    loc1, hl1, loc2, hl2 = args # mean, std dev, gaussian and expo amplitudes and half life
    ret  = 0
    mid  = 420
    ret  += np.concatenate((scipy.stats.moyal.pdf(x[x < mid] , loc = loc1 , scale = hl1) , np.zeros(len(x[x > mid])))) 
    ret  += np.concatenate((np.zeros(len(x[x < mid])) , scipy.stats.expon.pdf(x[x > mid] , loc = loc2 , scale = hl2))) 
    return ret
```

```python
def Delphes_custom(x, *args) :
    loc1, hl1, loc2, hl2, loc3, hl3 = args # mean, std dev, gaussian and expo amplitudes and half life
    ret    = 0
    break1 = 455
    break2 = 1100
    ret   += np.concatenate((scipy.stats.moyal.pdf(x[x < break1] , loc = loc1 , scale = hl1) , np.zeros(len(x[x > break1])))) 
    ret   += np.concatenate((np.zeros(len(x[x < break1])) , scipy.stats.expon.pdf(x[(x > break1) & (x < break2)] , loc = loc2 , scale = hl2), np.zeros(len(x[x > break2])) )) 
    ret   += np.concatenate((np.zeros(len(x[x < break2])) , scipy.stats.expon.pdf(x[ x > break2] , loc = loc3 , scale = hl3)))
    return ret
```

```python
def Delphes_custom1(x, *args) :
    loc1, hl1, loc2, hl2, loc3, hl3 = args # mean, std dev, gaussian and expo amplitudes and half life
    ret    = 0
    break1 = 455
    break2 = 1100
    if  (x < break1): 
        return scipy.stats.moyal.pdf(x, loc = loc1 , scale = hl1)
    elif(x > break1 and x < break2):
        return scipy.stats.expon.pdf(x, loc = loc2 , scale = hl2)
    elif(x > break2):
        return scipy.stats.expon.pdf(x, loc = loc3 , scale = hl3)
```

```python
def Run2_custom(x, *args) :
    
    loc1, hl1, loc2, hl2, loc3, hl3, loc4, hl4 = args # mean, std dev, gaussian and expo amplitudes and half life
    ret    = 0
    break1 = 455
    break2 = 1100
    break3 = 1800
    
    ret   += np.concatenate((scipy.stats.moyal.pdf(x[x < break1] , loc = loc1 , scale = hl1) , np.zeros(len(x[x > break1])))) 
    ret   += np.concatenate((np.zeros(len(x[x < break1])) , scipy.stats.expon.pdf(x[(x > break1) & (x < break2)] , loc = loc2 , scale = hl2), np.zeros(len(x[x > break2])) )) 
    ret   += np.concatenate((np.zeros(len(x[x < break2])) , scipy.stats.expon.pdf(x[(x > break2) & (x < break3)] , loc = loc3 , scale = hl3), np.zeros(len(x[x > break3])) )) 
    ret   += np.concatenate((np.zeros(len(x[x < break3])) , scipy.stats.expon.pdf(x[ x > break3] , loc = loc4 , scale = hl4)))
    return ret
```

```python
def Run2_custom1(x, *args) :
    loc1, hl1, loc2, hl2, loc3, hl3, loc4, hl4 = args # mean, std dev, gaussian and expo amplitudes and half life
    ret    = 0
    break1 = 455
    break2 = 1100
    break3 = 1800
    if  (x < break1): 
        return scipy.stats.moyal.pdf(x, loc = loc1 , scale = hl1)
    elif(x > break1 and x < break2):
        return scipy.stats.expon.pdf(x, loc = loc2 , scale = hl2)
    elif(x > break2 and x < break3):
        return scipy.stats.expon.pdf(x, loc = loc3 , scale = hl3)
    elif(x > break3) :
        return scipy.stats.expon.pdf(x, loc = loc4 , scale = hl4)
```
